```python
# Section 3: Project
def Takealot_scraper(url):
    """This function was made from the html of the webpage to be scraper 
    so it take in the html file for the webpage instead of going opnline since my internet 
    is unstable so url is the name of the html file from the webpage """
    # importing dependencies
    from bs4 import BeautifulSoup as bs
    import requests
    
    mydic={}
    mydic2 = {}
    
    # conecting to the url
    with open(url, "r", encoding="UTF-8") as f:
        
        soup = bs(f, "html.parser")
    
    # cheching the availability of the product
    in_stock = soup.find_all('div', class_ = "cell shrink stock-availability-status")
    availability = in_stock[0].span.contents[0]
    
        
    # Calculating the price of item
    contains_price = soup.find('div', class_="buybox-module_price_2YUFa")
    original_price = contains_price.span.contents[0]
    
    # finding the name of the seller
    seller_information = soup.find_all('div', class_="seller-information")
    
    # Calculating the next offer price of item
    next_offer = soup.find('div', class_="more-buying-choices-module_sf-more-buying-choices_vcCa6")
    next_available_price = next_offer.find_all("span")[1].contents[0]
    
    mydic["original_price"] = original_price
    mydic["availability"] = availability
    mydic["Seller"] = "Brand Company"
    mydic["next_offered_price"] = next_available_price
    
    

    
    if len(seller_information)== 0:
        return mydic
    else:
        
        Seller= seller_information[0].contents[0].a.contents[0]
        mydic2["original_price"] = original_price
        mydic2["availability"] = availability
        mydic2["Seller"] = Seller
        mydic2["next_offered_price"] = next_available_price
        return mydic2
    
```
